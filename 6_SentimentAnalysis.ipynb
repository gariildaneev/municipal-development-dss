# Сентимент анализ

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "blanchefort/rubert-base-cased-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
sentiment_model = AutoModelForSequenceClassification.from_pretrained(
    model_name
)
```

```python
sentiment_model.eval()
sentiment_model.to("cuda")
```

```python
import numpy as np

LABELS = ["negative", "neutral", "positive"]

def predict_sentiment(texts, batch_size=64):
    all_scores = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i + batch_size]

        inputs = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        ).to("cuda")

        with torch.no_grad():
            outputs = sentiment_model(**inputs)
            probs = torch.softmax(outputs.logits, dim=1)

        all_scores.append(probs.cpu().numpy())

    return np.vstack(all_scores)
```

```python
scores = predict_sentiment(df_clean["text"].tolist())
```

```python
df_clean["sent_neg"] = scores[:, 2]
df_clean["sent_neu"] = scores[:, 0]
df_clean["sent_pos"] = scores[:, 1]

df_clean["sentiment"] = np.argmax(scores, axis=1)
df_clean["sentiment_label"] = df_clean["sentiment"].map({
    2: "negative",
    0: "neutral",
    1: "positive"
})
```

```python
df_clean.to_csv('newsTopics&sentiment.csv', index=False)
```

# Zero-Shot Classification

```python
# возьмем небольшую подвыборку для теста
df_test = df_clean.sample(300, random_state=42).reset_index(drop=True)

texts = df_test["text"].tolist()
```

```python
remove_topics = [9, 14, 53, 55, 57, 66, 67, 68, 72, 86, 96, 98, 102, 105, 134, 152, 169, 211, 219, 241, 245, 252, 253, 271]
```

```python
df_clean_filtered = df_clean[~df_clean['topic'].isin(remove_topics)]
df_clean_filtered
```

```python
russian_interrogative_words = [
    "кто", "что", "когда", "где", "куда", "откуда",
    "почему", "зачем", "как", "какой", "чей", "сколько"
]

pattern = r'\b(?:' + '|'.join(russian_interrogative_words) + r')\b'

df_clean_filtered_no_questions = df_clean_filtered[
    ~df_clean_filtered['text'].str.contains(pattern, case=False, na=False)
]

df_clean_filtered_no_questions
```

```python
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli",
    device=0  # GPU
)
```

```python
LABELS = [
    "прогресс и достижения",
    "нейтральная информация",
    "кризис и проблемы"
]

CONF_THRESHOLD = 0.5
```

```python
texts = df_clean_filtered_no_questions["text"].tolist()
```

```python
from tqdm import tqdm

def zero_shot_policy_classification(
    texts,
    batch_size=8
):
    results = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i + batch_size]

        outputs = classifier(
            batch,
            LABELS,
            hypothesis_template="Этот новостной заголовок относится к категории {}."
        )


        for out in outputs:
            label = out["labels"][0]
            score = out["scores"][0]

            if score < CONF_THRESHOLD:
                label = "шум"

            results.append({
                "label": label,
                "score": score
            })

    return results
```

```python
import time

start = time.time()
zs_results = zero_shot_policy_classification(texts)
end = time.time()

print(f"Time: {(end - start):.2f} sec")
print(f"Avg per headline: {(end - start) / len(texts):.3f} sec")
```

```python
df_clean_filtered_no_questions["policy_label"] = [r["label"] for r in zs_results]
df_clean_filtered_no_questions["policy_score"] = [r["score"] for r in zs_results]
```
